## Empirical Cumulative Distribution Function

```python
def ecdf(data):
    #convert to numpy array
    if type(data) != 'numpy.ndarray':
        data = np.array(data)
    s_data = np.sort(data)
    u_data = np.unique(data)
    x_values = np.linspace(start = min(u_data),stop = max(u_data),num = len(u_data))
    y_values = []
    for val in x_values:
        temp = data[data <= val] # dont forget to change
        y_values.append(temp.size/len(data))
    return x_values,y_values
```

```python
x,y = ecdf(data['dem_share'])
```

```python
plt.plot(x,y,linestyle = 'none',marker = ".")
plt.show()
```
