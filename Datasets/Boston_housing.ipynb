```python
#importing libraries and Datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn import datasets
data = datasets.load_boston()
X = data.data
Y = data.target
data
```

```python
feature_names = data.feature_names
features = pd.DataFrame(X,columns = data.feature_names)
target = pd.DataFrame(Y)
```

```python
def draw_distplot(df,label,ax):
    _ = sns.distplot(df,ax = ax)
    ax.set_xlabel('{}'.format(label))
    ax.set_ylabel('Relative Distribution')
    ax.set_title('Distribution of {}'.format(label))
```

```python
f,axes = plt.subplots(7,2,figsize = (9,20))
axes = axes.ravel()
for i in range(len(feature_names)):
    draw_distplot(features[feature_names[i]],feature_names[i],ax = axes[i])
```

```python
data = pd.concat([features,target],axis = 1)
```

```python
corr = data.corr()
_ = sns.heatmap(corr,cmap = 'Greys')
```

```python
target_corr = corr.iloc[-1]
pos_corr = target_corr[target_corr >= 0]
neg_corr = target_corr[target_corr < 0]
```

```python
pos_corr_f = list(pos_corr.index)
neg_corr_f = list(neg_corr.index)
```
